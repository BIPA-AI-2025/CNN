```python
import torch
import torch.nn as nn
```

```python
class CNNModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 16, 3, stride=1, padding=1),  # [B, 1, 28, 28] → [B, 16, 28, 28]
            nn.ReLU(),
            nn.MaxPool2d(2),                           # → [B, 16, 14, 14]
            nn.Conv2d(16, 32, 3, stride=1, padding=1), # → [B, 32, 14, 14]
            nn.ReLU(),
            nn.MaxPool2d(2)                            # → [B, 32, 7, 7]
        )
        self.flatten = nn.Flatten()
        self.fc = nn.LazyLinear(10)  # 자동으로 in_features 계산

    def forward(self, x):
        x = self.conv(x)
        print("Before Flatten:", x.shape)  # [B, 32, 7, 7]

        x = self.flatten(x)
        print("After Flatten:", x.shape)   # [B, 1568] → 32 × 7 × 7 = 1568

        x = self.fc(x)
        return x
```

```python
# 테스트
model = CNNModel()
x = torch.randn(1, 1, 28, 28)
out = model(x)
print(out.shape)  # torch.Size([1, 10])
```
