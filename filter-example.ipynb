```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 이미지 경로 설정
input_path = 'elepant.jpeg'     # 예: 'input.jpg'
```

```python
# 1. 이미지 읽기 (BGR → GRAY)
img = cv2.imread(input_path)        # 예: 'input.jpg'
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# 수평 경계 검출 필터
v_filter_kernel = np.array([[1, 0, -1],
                          [1, 0, -1],
                          [1, 0, -1]], dtype=np.float32)
```

```python
# 수직 경계 검출 필터
h_filter_kernel = np.array([[ 1,  1,  1],
                          [ 0,  0,  0],
                          [-1, -1, -1]], dtype=np.float32)
```

```python
# 3. 필터 적용 (2D 컨볼루션)
filtered = cv2.filter2D(gray, ddepth=-1, kernel=v_filter_kernel)
```

```python
# 5. 결과 시각화 (원본 + 필터 결과)
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original (Gray)')
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Filtered Result')
plt.imshow(filtered, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
